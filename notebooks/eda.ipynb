### Union of happiness datasets

```python
def rename_columns(df, column_mapping):
    df.rename(columns=column_mapping, inplace=True)

# Define column mappings for each dataset
column_mapping1 = {
    'Country': 'country',
    'Region': 'region',
    'Happiness Rank': 'happiness_rank',
    'Happiness Score': 'happiness_score',
    'Economy (GDP per Capita)': 'gdp_per_capita',
    'Family': 'social_family_support',
    'Health (Life Expectancy)': 'health_life_expectancy',
    'Freedom': 'freedom',
    'Trust (Government Corruption)': 'perceptions_corruption',
    'Generosity': 'generosity',
    'Dystopia Residual': 'dystopia_residual'
}

column_mapping2 = {
    'Country': 'country',
    'Happiness.Rank': 'happiness_rank',
    'Happiness.Score': 'happiness_score',
    'Economy..GDP.per.Capita.': 'gdp_per_capita',
    'Family': 'social_family_support',
    'Health..Life.Expectancy.': 'health_life_expectancy',
    'Freedom': 'freedom',
    'Generosity': 'generosity',
    'Trust..Government.Corruption.': 'perceptions_corruption',
    'Dystopia.Residual': 'dystopia_residual'
}

column_mapping3 = {
    'Overall rank': 'happiness_rank',
    'Country or region': 'country',
    'Score': 'happiness_score',
    'GDP per capita': 'gdp_per_capita',
    'Social support': 'social_family_support',
    'Healthy life expectancy': 'health_life_expectancy',
    'Freedom to make life choices': 'freedom',
    'Generosity': 'generosity',
    'Perceptions of corruption': 'perceptions_corruption'
}
```

We use the "rename_columns" function to standardize the column names in the different datasets using a mapping dictionary. Since these datasets have different names in some columns, but these can be related in a single name. 

```python
# Rename the columns of the datasets (2015 and 2016)
rename_columns(df_2015, column_mapping1)
rename_columns(df_2016, column_mapping1)
```

For the 2015 and 2016 datasets we used the first column mapping as these retain the same structure and name of the data. 

```python
# Rename the columns of the dataset (2017)
rename_columns(df_2017, column_mapping2)
```

For the 2017 dataset we used the second column mapping since it has the names of its columns with dots, so we needed another dictionary to standardize the names. 

```python
# Rename the columns of the datasets (2018 y 2019)
rename_columns(df_2018, column_mapping3)
rename_columns(df_2019, column_mapping3)
```

For the 2018 and 2019 datasets we use the third column mapping as these retain the same structure and name of the data, in order to be standardized. 

```python
# Show column names to verify changes

print("2015 dataset:")
print(df_2015.columns)
```

```python
# Show column names to verify changes

print("2017 dataset:")
print(df_2017.columns)
```

```python
# Show column names to verify changes

print("2019 dataset:")
print(df_2019.columns)
```

```python
# Add the 'year' column to each dataset

df_2015['year'] = 2015
df_2016['year'] = 2016
df_2017['year'] = 2017
df_2018['year'] = 2018
df_2019['year'] = 2019
```

We create a year column for each of the datasets so that we can keep the data year when all datasets are collected. 

```python
# join the five datasets into one, keeping only the matching columns

df_happiness = pd.concat([df_2015, df_2016, df_2017, df_2018, df_2019], ignore_index=True, join="inner")
```

We do the concatenation of the 5 datasets in one which will have the same information of the columns of each dataset.

```python
# Show the first records of the combined dataset

print("Combined Happiness dataset:")
df_happiness
```

### EDA Happiness Dataset

```python
df_happiness.info()
```

The happiness dataset covers measurements of various aspects that impact well-being in each country, and extends from 2015 to 2019. Its columns include variables such as happiness score, gross domestic product per capita, family and social support, generosity, among others. With a total of 782 rows and 10 columns, this dataset presents a combination of three data types: float, int and object 

```python
# Nulls values per column

df_happiness.isnull().sum()
```

```python
# Delete rows with null values in the DataFrame
df_happiness.dropna(inplace=True)
```

The row containing null data in the corruption perception column is deleted. Since only one is not considered necessary imputation of data with the amount of records we have. 

```python
# Histogram of the happiness score

plt.figure(figsize=(8, 6))
sns.histplot(df_happiness['happiness_score'], bins=20, kde=True)
plt.title('Distribution of the Happiness Score')
plt.xlabel('Happiness Score')
plt.ylabel('frequency')
plt.show()
```

In the graph we can see that happiness scores are more frequent within the range of 4.5 to 6, while as happiness scores increase, the frequency of countries with higher scores decreases. 

```python
# Scatter Diagram: Happiness Score vs. GDP per Capita

plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_happiness, x='happiness_score', y='gdp_per_capita', hue='year')
plt.title('Happiness Score vs. GDP per Capita')
plt.xlabel('Happiness Score')
plt.ylabel('GDP per Capita')
plt.legend(title='Year')
plt.show()
```

In the graph, it can be seen that the lower happiness scores show a greater dispersion of data over the years, and these are associated with a lower gross domestic product. On the other hand, higher happiness scores are linked to higher per capita GDP, especially in the range of 1.25 to 1.75. 

```python
average_happiness_by_country = df_happiness.groupby('country')['happiness_score'].mean().reset_index()

top_10_countries = average_happiness_by_country.nlargest(10, 'happiness_score')

df_top_countries = df_happiness[df_happiness['country'].isin(top_10_countries['country'])]

# Bar chart for the 10 countries with the highest happiness scores


plt.figure(figsize=(10, 6))
sns.barplot(data=df_top_countries, x='country', y='happiness_score', hue='year', dodge=False)
plt.title('Happiness Score by Country (Top 10)')
plt.xlabel('Country')
plt.ylabel('Happiness Score')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Year')
plt.tight_layout()
plt.show()
```

Among the ten countries with the highest happiness scores, Finland ranks first with a score slightly above 7.50, according to 2019 data. The other countries maintain happiness scores above 7, although lower than Finland. 

```python
# exclude qualitative variables
df_numerico = df_happiness.drop(columns=['country'])

# Correlation Matrix

plt.figure(figsize=(10, 8))
corr = df_numerico.corr()
sns.heatmap(corr, annot=True, cmap='Blues', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

In the correlation matrix we can see the variables that are related to each other. Among this we have a high positive correlation between happiness score variables and gdp per capita of a value of 0.79, which indicates that if the gross domestic product of a country is higher the happiness score also increases. 

On the contrary, the happiness rank has large negative values in relation to almost all variables. It is not providing any useful information and will not be included.
